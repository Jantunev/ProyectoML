```python
import numpy as np
import pandas as pd
from scipy import stats as st
```

```python
df = pd.read_csv('data.csv', sep = ",", header=0, index_col=0)
```

```python
n_genes = df.shape[1]
```

```python
W = np.zeros((n_genes, n_genes))
P = np.zeros((n_genes, n_genes))
```

```python
df = df.to_numpy()[0:n_genes,0:n_genes]
```

```python
for i in range(0, n_genes):
    for j in range(0, n_genes):
        W[i,j], P[i,j] = st.spearmanr(df[:,i], df[:,j])
```

```python
np.savetxt("wadj_py.csv", W, delimiter = ",")
np.savetxt("pVal_py.csv", P, delimiter = ",")
```
