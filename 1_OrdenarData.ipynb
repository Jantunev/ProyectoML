Cargando la data por una única vez como archivos de trabajo, Si ya se habia descargado, imprime  Falso

```python
import os.path
```

```python
def Unica_Carga_Archivos(Nombre,dato):
    no_imprimir=os.path.isfile(Nombre)
    if not no_imprimir:
        dato.to_csv(Nombre,header=True,index=True)
    return(Nombre,not no_imprimir)
```

```python
Arch1,impr1=Unica_Carga_Archivos('df_carac_train.csv',X_train)
Arch2,impr2=Unica_Carga_Archivos('df_carac_test.csv',X_test)
Arch3,impr3=Unica_Carga_Archivos('df_y_train.csv',y_train)
Arch4,impr4=Unica_Carga_Archivos('df_y_test.csv',y_test)
```

```python
arreglo_df=pd.DataFrame(data=Arreglo,dtype=float).T
arreglo_df.columns=columnas
```

```python
arreglo_df.sort_values(by=[0],axis=1,ascending=False,inplace=True)
```

```python
def encontra_caract(df,b):
    c=len(df.columns)-1
    variacion=df.iloc[0,0:c].sum()
    print (variacion)
    valor=np.zeros(c,dtype=float)
    valor_div=np.zeros(c,dtype=float)
    valor[0]=df.iloc[0,0]
    for i in range(1,c):
        valor[i]=valor[i-1]+df.iloc[0,i]
        valor_div[i]=valor[i]/variacion
        #print (valor_div[i])
        if valor_div[i]>b:
            break
    print(i)
    return(valor,valor_div,i)
```

```python
Eval1,Eval2,num_carac=encontra_caract(arreglo_df,0.5)
```
