Separando la data de test, se realiza por una sola vez, Se prepara verificación si archivo ya existe

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df_caract,df_result,test_size=0.1,random_state=42)
```

```python
print("Tamaños de Conjuntos: \n X_train = ",X_train.shape,"\n X_test =", X_test.shape,"\n y_train =",y_train.shape,"\n y_test =",y_test.shape)
```

Cargando la data por una única vez como archivos de trabajo, Si ya se habia descargado, imprime  Falso

```python
import os.path
```

```python
def Unica_Carga_Archivos(Nombre,dato):
    no_imprimir=os.path.isfile(Nombre)
    if not no_imprimir:
        dato.to_csv(Nombre,header=True,index=True)
    return(Nombre,not no_imprimir)
```

```python
Arch1,impr1=Unica_Carga_Archivos('df_carac_train.csv',X_train)
Arch2,impr2=Unica_Carga_Archivos('df_carac_test.csv',X_test)
Arch3,impr3=Unica_Carga_Archivos('df_y_train.csv',y_train)
Arch4,impr4=Unica_Carga_Archivos('df_y_test.csv',y_test)
```

```python
print ('Archivos descargados \n', Arch1," ",impr1,"\n",Arch2,' ',impr2,"\n",Arch3,' ',impr3,'\n',Arch4,' ',impr4)
```
